```python
class ford_fulkerson:
    def __init__(self,node):
        self.node=node
        self.edge={}
        self.flow={}
        self.forward={}
        self.backward={}
        self.label={}
        for i in node:
            self.label[i]=False
        
        self.parent={}
        for i in node:
            self.parent[i]=None
            
    def add_edge(self,source,sink,capacity):
        if (source not in self.node) or (sink not in self.node):
            print('error')
            
        else:
            self.edge[(source,sink)]=capacity
            self.flow[(source,sink)]=0
            self.forward[(source,sink)]=capacity
            self.backward[(source,sink)]=0
            
    def find_augment_path(self,source,sink):
        I=[source]
        self.label[source]=True
        min_add=1000
        while True:
            if sink in I:
                break
            if len(I)==0:
                return None,0#means no augment path found
            
            a=I.pop(0)
            #for the case of forward arcs
            for e in self.edge.keys():
                if a==e[0]:
                    if self.forward[e]>0 and self.label[e[1]]==False:
                        #print(e[1])
                        I.append(e[1])
                        self.label[e[1]]=True
                        self.parent[e[1]]=a
                        if self.forward[e]<min_add:
                            min_add=self.forward[e]
                        
            #for the case of backward arc
            for e in self.edge.keys():
                if a==e[1]:
                    if self.backward[e]>0 and self.label[e[0]]==False:
                        #print(e[0])
                        I.append(e[0])
                        self.label[e[0]]=True
                        self.parent[e[0]]=a
                        if self.backward[e]<min_add:
                            min_add=self.backward[e]
            
        augment_path=[sink]
        t=self.parent[sink]
        #print(self.parent)
        while t!=None:
            #print(t)
            augment_path.append(t)
            t=self.parent[t]
        #print(min_add)

        augment_path=list(reversed(augment_path))
        return augment_path.copy(),min_add
        
    def find_max(self,source,sink):
        while True:
            augment_path,min_add=self.find_augment_path(source,sink)

            if augment_path==None:
                total=0
                for i in self.flow.keys():
                    print(str(i)+':',self.flow[i])
                    if i[0]==source:
                        total+=self.flow[i]

                print('total:%d'%total)
                break

            else:
                i=0
                while i<len(augment_path)-1:
                    a,b=augment_path[i],augment_path[i+1]
                    for e in self.edge.keys():
                        if a in e and b in e:
                            #for the case of fordward arc
                            if a==e[0]:
                                self.flow[e]+=min_add
                                self.forward[e]-=min_add
                                self.backward[e]+=min_add

                            #for the case of backward arc
                            else:
                                self.flow[e]-=min_add
                                self.forward[e]+=min_add
                                self.backward[e]-=min_add

                    i+=1
            #reset parent and label        
            for i in self.node:
                self.label[i]=False
            for i in self.node:
                self.parent[i]=None            
```
